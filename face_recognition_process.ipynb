```python
from face_recognition import test_images
from face_recognition import training_images
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
training_set_dir = 'datasets/train_english'
```

```python
fit_all_data = True 
```

```python
_, _, _, _ = training_images.train(training_set_dir,  
                                   image_size=144,
                                   process_output=False,
                                   all_data=fit_all_data, 
                                   clf_output=True,
                                   use_alignment=True, 
                                   language='english')
```

# Face Verification

```python
test_image_path = 'datasets/test/Verification/ChenZhenyu.jpg'
state = 'verification' 
used_labels = ['ChenZhenyu'] # face verification must choose one label
```

```python
answer, image_data = test_images.recognition(test_image_path,
                                             state=state,
                                             used_labels=used_labels,
                                             image_size=144,
                                             all_data=fit_all_data,
                                             output=False,
                                             language='english')

print(answer)
plt.figure(figsize=(12, 12))
plt.imshow(image_data)
plt.axis('off')
```
